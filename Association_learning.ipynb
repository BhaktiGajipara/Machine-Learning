```python
dataset = pd.read_csv(r"C:\Users\Bhakti Gajipara\Downloads\groceries - groceries.csv")
dataset.head(3)
```

```python
dataset = dataset.iloc[:,:-20]
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
market = []
for i in range(0,dataset.shape[0]):
    cus = []
    for j in dataset.columns:
        if type(dataset[j][i])== str:
            cus.append(dataset[j][i])
        market.append(cus)
```

```python
market
```

```python
l = []
for i in market:
    for j in i:
        l.append(j)
```

```python
l
```

```python
import collections
```

```python
p = collections.Counter(l)
```

```python
d = {"item name":p.keys(),"count":p.values()}
```

```python
pd.DataFrame(d).sort_values(by=["count"],ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
df = pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
df.head()
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df,min_support=0.05,use_colnames=True).sort_values(by = ["support"])
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fpgrowth(df,min_support=0.05,use_colnames=True,max_len=3).sort_values(by=["support"])
```
