```python
miss_var_per = cat_var.isnull().mean() *100
miss_var_per
```

```python
miss_var_per = cat_var.isnull().sum()/df.shape[0] *100
miss_var_per
```

```python
f = miss_var_per[miss_var_per>20].keys()
f
```

```python
cat_var.drop(columns = miss_var_per[miss_var_per>20].keys(),axis = 1,inplace =True)
cat_var
```

```python
cat_var.shape
```

```python
isnull_per = cat_var.isnull().mean()*100 
miss_var = isnull_per[isnull_per >0].keys()
miss_var
```

```python
cat_var["BsmtQual"].fillna("Missing")
```

```python
cat_var["BsmtQual"].mode()
```

```python
cat_var["BsmtQual"].value_counts()
```

```python
cat_var["BsmtQual"].fillna(cat_var["BsmtQual"].mode()[0])
```

```python
cat_var
```

```python
for var in miss_var:
    cat_var.fillna({var:cat_var[var].mode()[0]},inplace = True)
    print(var,"=",cat_var[var].mode()[0])
```

```python
cat_var.isnull().sum()
```

```python
pl.figure(figsize=(16,9))
for i,var in enumerate(miss_var):
    pl.subplot(2,5,i+1)
    pl.hist(cat_var[var],label = "imput")
    pl.hist(df[var].dropna(),label = "original")
    pl.legend()
```
