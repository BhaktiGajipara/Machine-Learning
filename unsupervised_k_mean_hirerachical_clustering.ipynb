# K-mean Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\Bhakti Gajipara\Downloads\Iris.csv")
dataset
```

```python
dataset = dataset.drop(columns=["Id","Species"])
```

```python
dataset.head(3)
```

```python
sns.pairplot(data = dataset)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,20):
    km = KMeans(n_clusters=i,init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)   # it give wcss value
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker = "o")
plt.xticks([i for i in range(2,21)])
plt.xlabel("no.of cluster")
plt.ylabel("wcss")
plt.grid(axis="x")
```

```python
from sklearn.metrics import silhouette_score
```

```python
kmn = KMeans(n_clusters=3)
dataset["predict"]= kmn.fit_predict(dataset)
```

```python
sns.pairplot(data = dataset,hue="predict")
```

```python
kmn.labels_
```

```python
silhouette_score(dataset,labels= kmn.labels_)
```

```python
a = []
no_c = [j for j in range(2,20)]
```

```python
for i in range(2,20):
    km1 = KMeans(n_clusters=i)
    km1.fit(dataset)
    a.append(silhouette_score(dataset,km1.labels_))
```

```python
plt.plot(no_c,a)
plt.xlabel("no. of cluster")
plt.ylabel("silhouette score")
plt.xticks(no_c)
plt.grid(axis="x")
```

# hirerachical clustering

```python
dataset1 = pd.read_csv(r"C:\Users\Bhakti Gajipara\Downloads\Iris.csv")
dataset1
```

```python
dataset1 = dataset1.drop(columns=["Id","Species"])
```

```python
import scipy.cluster.hierarchy as sc
```

```python
sc.dendrogram(sc.linkage(dataset1,method = "single",metric = "euclidean"))
plt.show()
```
