```python
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\Users\Bhakti Gajipara\Downloads\Social_Network_Ads.csv")
dataset
```

```python
dataset.drop(columns=["User ID","Gender"],inplace = True)
```

```python
dataset.isnull().sum()
```

```python
sns.kdeplot(data = dataset["Age"])
```

```python
sns.scatterplot(x = "Age",y = "EstimatedSalary",data = dataset,hue = "Purchased")
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test),gnb.score(x_train,y_train)
```

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test),mnb.score(x_train,y_train)
```

```python
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test),bnb.score(x_train,y_train)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = gnb)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = mnb)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = bnb)
```
