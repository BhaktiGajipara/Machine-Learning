```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\Bhakti Gajipara\OneDrive\Documents\machine learing\Ice_cream selling data.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
plt.scatter(dataset["Temperature (°C)"],dataset["Ice Cream Sales (units)"])
plt.show()
```

```python
dataset.corr()
```

```python
x = dataset[["Temperature (°C)"]]
y = dataset[["Ice Cream Sales (units)"]]
```

```python
from sklearn.preprocessing import PolynomialFeatures 
```

```python
pf = PolynomialFeatures(degree=3)
pf.fit(x)
```

```python
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=53)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)    # it show accuracy of  machine algorithm
```

```python
test = pf.transform([[-4.213985]])
```

```python
lr.predict(test)
```

```python
prd = lr.predict(x)
```
