# k-nearest neighbours in classification

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\Users\Bhakti Gajipara\OneDrive\Documents\machine learing\placement.csv")
dataset
```

```python
x = dataset.iloc[:,:-1]
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
kn = KNeighborsRegressor(n_neighbors=15)
kn.fit(x_train,y_train)
```

```python
kn.score(x_test,y_test)
```

```python
kn.score(x_train,y_train)
```

```python
for i in range(1,30):
    knn1 = KNeighborsRegressor(n_neighbors=i)
    knn1.fit(x_train,y_train)
    print(knn1.score(x_train,y_train),knn1.score(x_test,y_test),i)
```
