# Logistic Regression And Binary classification

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Bhakti Gajipara\Downloads\Social_Network_Ads.csv")
dataset.drop(columns=["User ID","Gender","EstimatedSalary"],inplace= True)
```

```python
dataset
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="Age",y = "Purchased",data = dataset)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="Age",y = "Purchased",data = dataset)
sns.lineplot(x = "Age",y = lr.predict(x),data = dataset)
```

# Logistic Regression And Binary classification, Multiple Input

```python
df = pd.read_csv(r"C:\Users\Bhakti Gajipara\Downloads\Social_Network_Ads.csv")
df.head()
```

```python
df.drop(columns=["User ID","Gender"],inplace= True)
```

```python
df
```

```python
sns.scatterplot(x = "Age",y = "EstimatedSalary",data = df,hue="Purchased")
```

```python
x1 = df.iloc[:,:-1]
y1= df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x1_train,y1_train)
```

```python
lr.score(x1_test,y1_test)*100
```

```python
sns.scatterplot(x = "Age",y = "EstimatedSalary",data = df,hue="Purchased")
sns.lineplot(x = "Age",y = "EstimatedSalary",data = df,hue=lr.predict(x1))
```

```python
lr.predict([[22,22000]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plt.figure(figsize=(5,5))
plot_decision_regions(x1.to_numpy(),y1.to_numpy(),clf = lr)
plt.show()
```
