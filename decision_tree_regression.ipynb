# decision tree in regression

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor,plot_tree
```

```python
dataset = pd.read_csv(r"C:\Users\Bhakti Gajipara\OneDrive\Documents\machine learing\Salary_Data.csv")
dataset
```

```python
sns.pairplot(data = dataset)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_teat,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
dt = DecisionTreeRegressor(max_depth=3)
dt.fit(x_train,y_train)
```

```python
dt.score(x_teat,y_test)
```

```python
dt.score(x_train,y_train)
```

```python
plt.figure(figsize=(15,15))
plot_tree(dt)
plt.show()
```
