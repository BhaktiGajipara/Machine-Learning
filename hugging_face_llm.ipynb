```python
import dataset
import transformers
import huggingface_hub
import matplotlib.pyplot as plt
```

```python
from transformers import pipeline
classifier = pipeline("text-classification")
```

```python
classifier("text")
```

```python
text1 = "Narendra Modi is prime minister of India"
```

```python
import pandas as pd
```

# NER - (Name Entity Recognition)

```python
ner = pipeline("ner",aggregation_strategy="simple")
out =  ner(text1)
pd.DataFrame(out)
```

# Question Answering

```python
reader = pipeline("question-answering")
```

```python
question = "What user want?"
```

```python
output = reader(question=question, context=text)
pd.DataFrame([output])
```

# summerization

```python
summerize = pipeline("summarization")
```

```python
summry = summerize(text,clean_up_tokenization_spaces=True,max_length=130)
summry[0]['summary_text']
```

# Translation

```python
translation = pipeline("translation_en_to_de",model="Helsinki-NLP/opus-mt-en-fr")
```

```python
output = translation(text,clean_up_tokenization_spaces=True)
output[0]['translation_text']
```
