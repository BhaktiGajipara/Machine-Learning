```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\Bhakti Gajipara\OneDrive\Documents\machine learing\placement.csv")
dataset.head()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train),lr.score(x_test,y_test)
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt.score(x_train,y_train),dt.score(x_test,y_test)
```

```python
sv = SVR()
sv.fit(x_train,y_train)
sv.score(x_train,y_train),sv.score(x_test,y_test)
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("lr1",LinearRegression()),("dt1",DecisionTreeRegressor()),("sv1",SVR())]
```

```python
vc = VotingRegressor(li,weights=[60,20,20])
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train) , vc.score(x_test,y_test)
```

```python
prd = {"dt":dt.predict(x_test),"svm":sv.predict(x_test),"lr":lr.predict(x_test),"predict":vc.predict(x_test)}
```
