```python
missing_value_per = df.isnull().sum() / df.shape[0] * 100
missing_value_per
```

```python
missing_var_clm_gtr_20 = missing_value_per[missing_value_per > 20].keys()
missing_var_clm_gtr_20
```

```python
df2_drop_clm = df.drop(columns = missing_var_clm_gtr_20)
df2_drop_clm
```

```python
df3_num = df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
df3_num[df3_num.isnull().any(axis = 1)]
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum() > 0]
missing_num_var
```

```python
pl.figure(figsize=(25,25))
sns.set()
for i,var in enumerate(missing_num_var):
    pl.subplot(9,4,i+1)
    sns.distplot(df3_num[var],bins = 20,kde_kws={'linewidth':5,'color':'#DC143C'})
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
pl.figure(figsize=(25,25))
for i,var in enumerate(missing_num_var):
    pl.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins = 20,label ="original")
    sns.distplot(df4_num_mean[var],bins = 20,label ="Mean")
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
pl.figure(figsize=(25,25))
for i,var in enumerate(missing_num_var):
    pl.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins = 20,label ="original")
    sns.distplot(df4_num_mean[var],bins = 20,label ="mean")
    sns.distplot(df5_num_median[var],bins = 20,label ="Median")
```

```python
df_concate = pd.concat([df3_num[missing_num_var],df4_num_mean[missing_num_var],df5_num_median[missing_num_var]],axis = 1)
```
